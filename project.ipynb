```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python

df
```

```python
print("\nDataset Shape:", df.shape)
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python




                                                        #DATA CLEANING
```

```python
df['country'] = df['country'].fillna('Unknown')
```

```python
df['director'] = df['director'].fillna('Not Given')
```

```python
df.dropna(subset=['rating', 'duration'], inplace=True)  # DROPPING ROWS WITH MISSING RATING
```

```python
df
```

```python

 
                                                    # EDA ANALYSIS
```

```python
#  1. MOVIES VS TV SHOWS




type_counts = df['type'].value_counts()
```

```python
sns.set(style="whitegrid")
plt.figure()
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Movies vs TV Shows Distribution")
plt.ylabel("")
plt.show()
```

```python
# 2. Content added over years


year_counts = df['release_year'].value_counts().sort_index()
```

```python
plt.figure()
sns.boxplot(x='type', y='release_year', data=df)
plt.title("Release Year Distribution by Content Type")
plt.show()
```

```python
# 3. Top 10 content producing countries



country_counts = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
country_counts.sort_values().plot(kind='barh')
plt.title("Top 10 Content Producing Countries on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python

# 4. Ratings distribution


rating_counts = df['rating'].value_counts()
rating_counts = df['rating'].value_counts()

plt.figure(figsize=(10,6))
rating_counts.sort_values().plot(kind='barh')
plt.title("Distribution of Netflix Ratings")
plt.xlabel("Number of Titles")
plt.ylabel("Rating")
plt.show()



```

```python

# 5. Top 10 genres



genre_counts = df['listed_in'].value_counts().head(10)



```

```python
plt.figure(figsize=(8,8))
genre_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Top 10 Genres Distribution on Netflix")
plt.ylabel("")
plt.show()
```
